# Performing EDA

```python
fig, ((a1, a2), (a3, a4)) = plt.subplots(nrows=2,ncols=2,figsize=(20, 10),gridspec_kw={'width_ratios': [2,1]})

sns.countplot(data=df, x='Region', ax=a1, color='green')
a1.set_title('Count of Regions')

sns.countplot(data=df, x='Income', ax=a2)
a2.set_title('Count of Income types')

sns.barplot(data=df, x='Region', y='Population', ax=a3, color='green')
a3.set_title('Population by Region')

sns.barplot(data=df, x='Income', y='Population', ax=a4)
a4.set_title('Population by Income types')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data=df, x='Region', y='Population', estimator=sum)
plt.title('Population by Region (based on sum of values)')
plt.xlabel('Region')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Income', y='Population', data=df, estimator=sum)
plt.title('Total Population by Income Level (based on sum of values)')
plt.xlabel('Income Level')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(50, 6))
sns.barplot(x='Year', y='Population', data=df, estimator=sum)
plt.title('Total Population by Year')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='Region', y='Population', hue='Income', data=df, estimator=sum)
plt.title('Total Population by Region and Income Level')
plt.xlabel('Region')
plt.ylabel('Total Population')
plt.yscale('log')
plt.legend(title='Income Level')
plt.show()
# some gaps are there as the values are very low!
```

```python
g = sns.displot(
    data=df,
    x='Population',
    col='Region',
    hue='Region',
    kind='hist',
    kde=True,
    log_scale=True)
g.set_titles(col_template='{col_name} Region')
plt.show()
```

```python
g = sns.displot(
    data=df,
    x='Population',
    col='Income',
    hue='Income',
    kind='hist',
    kde=True,
    log_scale=True)
g.set_titles(col_template='{col_name} Income Level')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Population'], bins=10, kde=True, log_scale=True)
plt.title('Distribution of Population (Log Scale)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
